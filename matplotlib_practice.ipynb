```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axia to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May","Jun", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the Plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Farce($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis,y_axis, marker="*", color='blue', linewidth=2, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
fig, ax = plt.subplots()
# Change the line color to green, make the marker, change teh width to 2 points 
ax.plot(x_axis, y_axis, marker= "*", color="green", linewidth=2, label='Boston')
# Set titles for the x and y axis
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")
# Set the limit for y
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#change the bar color to magenta
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
```

```python
plt.plot(y_axis,x_axis,'o',color='red',label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.legend()
plt.show()

```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
plt.scatter(x_axis,y_axis, s = [i*3 for i in y_axis])
```

```python
# Create the plot
fig, ax = plt.subplots()
# Change the line color to green, make the marker, change teh width to 2 points 
ax.scatter(y_axis,x_axis, marker= "o", color="blue",edgecolor='k',label='Boston', linewidth=2, alpha = 0.20, s=[i*5 for i in y_axis])

# Set titles for the x and y axis
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0,50)
# Set the limit for y
ax.set_title("PyBer Fare by Month")
ax.legend()


```

```python
plt.pie(y_axis,labels=x_axis)
plt.show
```

```python
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# object oriented
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,explode_values, colors= colors, labels=x_axis, shadow = True, counterclock=False, autopct='%.1f%%',
      startangle = 86)
plt.show()
```
